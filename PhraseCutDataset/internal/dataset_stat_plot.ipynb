```python
import json
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

import os
os.chdir('/mnt/ilcompf8d1/user/chenwu/PhraseCut/')
print(os.getcwd())
```

```python
with open('data/refvg/amt_result/name_att_rel_count_amt.json', 'r') as f:
    c_d = json.load(f)
names = c_d['name']
atts = c_d['att']
rels = c_d['rel']
```

```python
n_s = sorted([kv for kv in names.items()], key=lambda kv:-kv[1])
a_s = sorted([kv for kv in atts.items()], key=lambda kv:-kv[1])
r_s = sorted([kv for kv in rels.items()], key=lambda kv:-kv[1])
```

```python
fig, axs = plt.subplots(3,1, figsize=[13, 10])

def autolabel(rects, ax, max_h):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        h = rect.get_height()
        height = min(rect.get_height(), max_h*0.8)
        ax.text(rect.get_x() + rect.get_width()/2., height +3,
                '%d' % int(h),
                ha='center', va='bottom',fontsize=8, rotation=-45)

end = 1272
gap = 20
n = n_s[0: end+1: gap]
nl = ['%d:%s' % (i * gap, a[0]) for i, a in enumerate(n)]
nc = [a[1] for a in n]        
        
rects = axs[0].bar(np.arange(len(nl)), nc, alpha=0.5)
autolabel(rects, axs[0], 1000)
axs[0].set_xticks(np.arange(len(nl)))
axs[0].set_xticklabels(nl, rotation=-90,horizontalalignment='center')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Name Frequency')
axs[0].set_ylim([0,1000])


end = 593
gap = 10
n = a_s[0: end+1: gap]
nl = ['%d:%s' % (i * gap, a[0]) for i, a in enumerate(n)]
nc = [a[1] for a in n]        
        
rects = axs[1].bar(np.arange(len(nl)), nc, alpha=0.5)
autolabel(rects, axs[1], 800)
axs[1].set_xticks(np.arange(len(nl)))
axs[1].set_xticklabels(nl, rotation=-90,horizontalalignment='center')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Attribute Frequency')
axs[1].set_ylim([0,800])

end = 126
gap = 2
n = r_s[0: end+1: gap]
nl = ['%d:%s' % (i * gap, a[0]) for i, a in enumerate(n)]
nc = [a[1] for a in n]        
        
rects = axs[2].bar(np.arange(len(nl)), nc, alpha=0.5)
autolabel(rects, axs[2], 800)
axs[2].set_xticks(np.arange(len(nl)))
axs[2].set_xticklabels(nl, rotation=-90,horizontalalignment='center')
axs[2].set_ylabel('Frequency')
axs[2].set_title('Relationship Description Frequency')
axs[2].set_ylim([0,800])



fig.tight_layout() 
plt.show()
```

```python
with open('data/refvg/scene_graphs_pp.json', 'r') as f:
    sg = json.load(f)
```

```python
with open('data/refvg/amt_result/refer_filtered_instance.json', 'r') as f:
    refer = json.load(f)
```

```python
Imgs_info = {img['image_id'] : img for img in imgs_info}
    
```

```python
object_dict = dict()
for img in sg:
    for obj in img['objects']:
        object_dict[obj['object_id']] = obj
```

```python
from utils.visualize import visualize_refvg
import random

img_count = 40
fig, axes = plt.subplots(img_count / 2, 2, figsize=(6, 1.2 * img_count))

sub_refer = random.sample(refer, img_count)
for i, r in enumerate(sub_refer):
    print('img %d/%d: %d' % (i, img_count, r['image_id']))

    vg_objs = []
    for ann_id in r['ann_ids']:
        if ann_id not in object_dict:
            print('missing obj:', r['image_id'], ann_id)
            continue
        vg_objs.append(object_dict[ann_id])
    vg_boxes = [[obj['x'], obj['y'], obj['w'], obj['h']] for obj in vg_objs]
    
    name = None
    atts = None
    rels = None
    tp = None
    if r['phrase_structure']:
        name = r['phrase_structure']['name']
        atts = r['phrase_structure']['attributes']
        rels = [rel['predicate'] for rel in r['phrase_structure']['relations']]
        tp = r['phrase_structure']['type']
    version = Imgs_info[r['image_id']]['refvg_version']
    title = '%s: %s \n[%s: %s|%s|%s]' % (version, r['phrase'] , str(tp), str(name),str(atts), str(rels))
#     print title
#     title = r['phrase']
    if 'Polygons' in r:
        r['polygons'] = [p for ins_ps in r['Polygons'] for p in ins_ps]

    visualize_refvg(ax=axes.flatten()[i], title=title, img_id=r['image_id'],gt_polygons=r['polygons'],
                    gt_boxes=r['instance_boxes'], vg_boxes=vg_boxes)

# plt.subplots_adjust(left=0.1, right=0.9, top=0.99, bottom=0.01)
fig.tight_layout()
```
